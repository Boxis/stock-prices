```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))  # so `import stocks` works from examples/
%matplotlib inline

from stocks import (
    get_prices, get_current, compute_performance,
    flag_big_moves, plot_prices,
)
```

## 1. Fetch historical prices

Pass a single ticker or a list. Dates are `YYYY-MM-DD`; `end=None` means today.

```python
tickers = ['AAPL', 'MSFT', 'NVDA']
prices = get_prices(tickers, start='2024-01-01', end='2024-06-28')
prices.head()
```

## 2. Performance summary (one row per ticker)

```python
perf = compute_performance(prices)
perf
```

## 3. Charts

Raw adjusted close, and a normalized (rebased-to-100) comparison.

```python
plot_prices(prices, normalize=False, show=True)
plot_prices(prices, normalize=True, show=True)
```

## 4. Largest daily moves and likely news drivers

```python
moves = flag_big_moves(prices, top_n=5)
moves[['ticker', 'date', 'daily_return', 'headline', 'publisher', 'url']]
```
